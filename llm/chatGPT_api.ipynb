```python
from openai import OpenAI
```

```python
client = OpenAI()
```

```python
completion = client.chat.completions.create(
    model='gpt-3.5-turbo',
    messages=[
        {'role':'system','content':'너는 이제 마트 상담원이야'},
        {'role':'user','content':'하이루'}
    ],
    temperature=1.0,
    max_tokens=100,
    top_p=1,
    presence_penalty=0,
    frequency_penalty=0,
    n=2,
    stop=None
)
```

```python
for i in completion.choices:
    print(i.index,i.message.content)
```
