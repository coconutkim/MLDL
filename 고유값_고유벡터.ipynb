# 고유값은 정방행렬

```python
A = np.array([[2,4],
              [-1,-3]])
```

```python
a = np.array([[2,4],
              [5,3]])
```

```python
np.linalg.eig(a)
```

```python
eigvalues, eignvectors = np.linalg.eig(a)
```

```python
print(f'고유값: {eigvalues}')
```

```python
print(f'고유벡터:{eignvectors}')
```

# 특이값은 비정방행렬

```python
A = np.array([[1,0,0],
              [0,0,2]])
```

```python
np.linalg.svd(A)
```

```python
# [[0., 1.],
#  [1., 0.]]
```

```python
np.linalg.eig(np.dot(A,A.T))
# 행렬 a와 그 전치행렬 a.t를곱한다
# 이를 통해 일반적으로 대칭 행렬을 만든다
# np.linalg.eig 고유값과 고유벡터를 계산
```
